# 3. bfill()

a. Create a DataFrame with some missing values and fill them using bfill().

```python
df=pd.DataFrame(data = {'Name': ['chekku', 'chinnu', 'kichu'],
                        'Age': [9,None,6],
                        'Mark':[7.5,8.3,2.5]})

df.bfill()
```

b. Demonstrate how bfill() works with multiple columns in a DataFrame.

```python
df=pd.DataFrame(data = {'Name': ['chekku', 'chinnu', 'kichu','abhi','vishnu'],
                        'Age': [9,None,6,None,9],
                        'Mark':[7.5,8.3,2.5,None,3.5]})

df.bfill()
```

c. Show how to limit the number of bfill() operations to one row.

```python
df=pd.DataFrame(data = {'Name': ['chekku', None, 'kichu','abhi','vishnu'],
                        'Age': [9,None,6,None,9],
                        'Mark':[7.5,8.3,2.5,None,3.5]})
print(df)
print('\n')

df_bfilled=df.bfill(limit=1)
df_bfilled
```

# 5. combine()

a. Combine two DataFrames by comparing each element and selecting the
larger element.

```python
import numpy as np

df1 = pd.DataFrame([[1,2,3,4,5],[6,7,8,9,0]])

df2 = pd.DataFrame([[0,9,8,7,6],[5,4,3,2,1]])

df1.combine(df2, np.maximum)
# df1.combine(df2, lambda a, b: np.where(a > b, a, b))
```

b. Create a function that performs a custom operation while combining two
DataFrames.

```python
import numpy as np

def myfunc(a, b):
    return np.where(a > b, a - b, b - a)


df1 = pd.DataFrame([[1,2,3,4,5],[6,7,8,9,0]])

df2 = pd.DataFrame([[0,9,8,7,6],[5,4,3,2,1]])

df1.combine(df2, myfunc)
```

c. Combine two DataFrames column-wise and handle NaN values using a
custom function.

```python
df1 = pd.DataFrame([[1,None,3,None,5],[None,7,None,9,None]])

df2 = pd.DataFrame([[None,9,None,7,None],[5,None,3,None,1]])
df1.combine_first(df2)
```

# 8. copy()

a. Create a deep copy of a DataFrame and modify it without affecting the
original.

```python
df = pd.DataFrame({'Name':['pp','chekku'],
      'Rollno':[4,5],
      'Mark':[60,50]})
df1 = df.copy(deep=True)
df1.loc[0,'Rollno']=1
print(df)
print('\n')
print(df1)
```

b. Make a shallow copy of a DataFrame and observe how changes affect both.

```python
df = pd.DataFrame({'Name':['pp','chekku'],
      'Rollno':[4,5],
      'Mark':[60,50]})
df1 = df.copy(deep=False)
df1.loc[0,'Rollno']=1
print(df)
print('\n')
print(df1)
```

c. Modify a DataFrame after using copy(deep=False) and explain the
behavior.

```python
df = pd.DataFrame({'Name':['pp','chekku'],
      'Rollno':[4,5],
      'Mark':[60,50]})
df1 = df.copy(deep=False)
df1.loc[0,'Rollno']=1
print(df)
print('\n')
print(df1)
```

# 9. cummax()

a. Compute the cumulative maximum of a column in a DataFrame.

```python
df = pd.DataFrame({'Name':['pp','chekku'],
      'Rollno':[4,5],
      'Mark':[60,50]})

df['Rollno'].cumsum()
```

b. Apply cummax() row-wise across multiple columns.

```python
df = pd.DataFrame({'Rollno':[4,5],
      'Mark':[60,50]})
df.cumsum(axis=1)
```

c. Use cummax() to find the cumulative maximum of a Series containing both
positive and negative values.

```python
ser = pd.Series([1,-4,-6,9,3,7])
ser.cumsum()
```
