18. Create a DataFrame with duplicate rows. Remove all duplicates using
the drop_duplicates() method.

```python
data = {
  "Duration": [50, 40, 40, 45],
  "Pulse": [109, 117, 117, 110],
  "Calories": [409, 479, 479, 340]
}
df=pd.DataFrame(data)
df.drop_duplicates(inplace=True)
df
```

19. Write a function to clean a DataFrame by filling missing data, fixing
incorrect types, and removing duplicates.

```python
def clean_df(df):
  df=df.fillna(df.mean())
  df=df.drop_duplicates()
  df=df.astype(float)
  return df

data = {
  "Duration": [50, None, 45],
  "Pulse": [109, 117, 110],
  "Calories": [None, 479.5, 340.5]
}
df=pd.DataFrame(data)
print(clean_df(df))
```

# Data Correlation:

# Aggregation Functions:

21. Create a DataFrame and use the agg() method to calculate the sum
and mean of numeric columns.

```python
data = {
  "Duration": [50, 40, 45],
  "Pulse": [109, 117, 110],
  "Calories": [409, 479, 340]
}
df=pd.DataFrame(data)
x=df.agg(['sum','mean'])

x
```

22. Create a DataFrame and use the agg() function to apply multiple
aggregation operations (e.g., min, max, mean) on a specific column.

```python
data = {
  "Duration": [50, 40, 45],
  "Pulse": [109, 117, 110],
  "Calories": [409, 479, 340]
}
df=pd.DataFrame(data)
x=df['Duration'].agg(['min','max','mean'])
# y=df.agg({'Duration' : ['min','max','mean']})
x
```

23. Create a DataFrame and use the agg() method with a custom
aggregation function (e.g., calculate the range).

```python
def custom(df):
  return df.max() - df.min()

data = {
  "Duration": [50, 40, 45],
  "Pulse": [109, 117, 110],
  "Calories": [409, 479, 340]
}
df=pd.DataFrame(data)

print(df.agg({'Calories':custom}))

print('\n')

print(df.agg(custom))
```
